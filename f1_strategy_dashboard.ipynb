```

import streamlit as st
import fastf1
import matplotlib.pyplot as plt
import numpy as np

fastf1.Cache.enable_cache('cache')

st.title("F1 Strategy Dashboard")

def load_session(year, gp):
    session = fastf1.get_session(year, gp, 'R')
    session.load()
    return session

def visualize_pit_stops(session):
    pit_stops = session.get_pitstops()
    fig, ax = plt.subplots()
    for driver, stops in pit_stops.groupby('Driver'):
        ax.scatter(stops['Lap'], stops['Stop'], label=driver)
    st.pyplot(fig)

session = load_session(2023, 'Japan')
visualize_pit_stops(session)
```
